```python
import joblib
from utils.preprocessing import clean_text

# Load v1 model
lr_v1 = joblib.load("../models/Sentiment_Classifier_v1/lr_model.joblib")
tfidf_v1 = joblib.load("../models/Sentiment_Classifier_v1/tfidf_vectorizer.joblib")
```

```python
# Test cases
reviews = [
    "I absolutely loved this movie! The acting was amazing and the story was gripping.",
    "This was the worst movie I have ever seen. Totally boring and a waste of time."
]

# Clean the reviews
cleaned = [clean_text(r) for r in reviews]

# Transform and predict
X_test = tfidf_v1.transform(cleaned)
predictions = lr_v1.predict(X_test)

# Map 0/1 back to sentiment
sentiments = ["negative" if p==0 else "positive" for p in predictions]

for r, s in zip(reviews, sentiments):
    print(f"Review: {r}\nPredicted Sentiment: {s}\n")
```
